```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Part 1 => BACKGROUND RADIATION
df_back_ground = pd.read_csv('/home/nkk/Área de trabalho/projetos/IC/back_ground.csv', sep=',')
df_back_ground
```

```python
# Estatisticas descritivas
print(df_back_ground.describe())
```

```python
print(f'Média - CPM: {df_back_ground['CPM'].mean():.4f}')
print(f'Mediana - CPM: {df_back_ground['CPM'].median():.4f}')
print(f'Desvio do Padrão - CPM: {df_back_ground['CPM'].std():.4f}') 
```

```python
plt.xlabel('CPM')
plt.ylabel('Frequência')
plt.plot(df_back_ground['CPM'], label='Background')
plt.show()
```
